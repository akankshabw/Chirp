```python
def plot_spectrogram(y, sr, ax, title):
    D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log', ax=ax)
    ax.set_title(title)


fig, ax = plt.subplots(5, 2, figsize=(10, 25))
fig.tight_layout(pad=5.0)

# Iterate over each pair of x and y
for i in range(5):
    onset_env_x = librosa.onset.onset_strength(y=locals()[f'x{i+1}'], sr=locals()[f'sr_{i+1}'])
    onset_env_y = librosa.onset.onset_strength(y=locals()[f'y{i+1}'], sr=locals()[f'sr{i+1}'])

    plot_spectrogram(locals()[f'x{i+1}'], locals()[f'sr_{i+1}'], ax[i, 0], f"Spectrogram {i+1} - x{i+1}")
    plot_spectrogram(locals()[f'y{i+1}'], locals()[f'sr{i+1}'], ax[i, 1], f"Spectrogram {i+1} - y{i+1}")

plt.suptitle("Spectrogram graphs")
plt.show()
```
