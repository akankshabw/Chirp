```python
import librosa # for audio analysis
import librosa.display
import soundfile as sf # for reading and writing sound files
import matplotlib.pyplot as plt
import IPython.display as ipd # to display interactive rich media like audios
import numpy as np
```

```python
audio5_1 = "/content/5_yweweler_10.wav"
audio5_2 = "/content/5_yweweler_11.wav"
audio5_3 = "/content/5_yweweler_2.wav"
audio9_1 = "/content/9_yweweler_4.wav"
audio9_2 = "/content/9_yweweler_5.wav"
audio9_3 = "/content/9_yweweler_6.wav"
```

Load all the audio files of number 5.


```python
y1, sr1 = librosa.load(audio5_1)
y2, sr2 = librosa.load(audio5_2)
y3, sr3 = librosa.load(audio5_3)
```

Load all the audio files of the number 9.

```python
x1, sr_1 = librosa.load(audio9_1)
x2, sr_2 = librosa.load(audio9_2)
x3, sr_3 = librosa.load(audio9_3)
```

Play the audio of any one of the three files of each number to ensure that everything is fine tiil now.

```python
print("Audio 9:")
ipd.Audio(x1, rate=sr_1)
```

```python
print("Audio 5:")
ipd.Audio(y3, rate=sr3)
```

```python
mfcc1 = librosa.feature.mfcc(y=y1, sr=sr1)
mfcc2 = librosa.feature.mfcc(y=y2, sr=sr2)
mfcc3 = librosa.feature.mfcc(y=y3, sr=sr3)
print("mfcc1:", np.max(mfcc1),np.min(mfcc1))
print("mfcc2:", np.max(mfcc2),np.min(mfcc2))
print("mfcc3:", np.max(mfcc3),np.min(mfcc3))
```

```python
mfcc1 = librosa.feature.mfcc(y=x1, sr=sr_1)
mfcc2 = librosa.feature.mfcc(y=x2, sr=sr_2)
mfcc3 = librosa.feature.mfcc(y=x3, sr=sr_3)
print("mfcc1:", np.max(mfcc1),np.min(mfcc1))
print("mfcc2:", np.max(mfcc2),np.min(mfcc2))
print("mfcc3:", np.max(mfcc3),np.min(mfcc3))
```

Not much difference in ranges.

```python
#Zero Crossing Rate of number 5
fig, ax = plt.subplots(1, 3,figsize=(15,4))
fig.tight_layout(pad=5.0)

zcr1 = librosa.feature.zero_crossing_rate(y1)
plt.subplot(1, 3, 1)
plt.plot(zcr1[0], linestyle = 'solid');

zcr2 = librosa.feature.zero_crossing_rate(y2)
plt.subplot(1, 3, 2)
plt.plot(zcr2[0],linestyle = 'solid');

zcr3 = librosa.feature.zero_crossing_rate(y3)
plt.subplot(1, 3, 3)
plt.plot(zcr3[0], linestyle = 'solid');
plt.suptitle("Zero Crossing Rate graphs for the number 5")
plt.show();
```

```python
#Zero Crossing Rate of number 9
fig, ax = plt.subplots(1, 3,figsize=(15,4))
fig.tight_layout(pad=5.0)

zcr1 = librosa.feature.zero_crossing_rate(x1)
plt.subplot(1, 3, 1)
plt.plot(zcr1[0], linestyle = 'solid');

zcr2 = librosa.feature.zero_crossing_rate(x2)
plt.subplot(1, 3, 2)
plt.plot(zcr2[0],linestyle = 'solid');

zcr3 = librosa.feature.zero_crossing_rate(x3)
plt.subplot(1, 3, 3)
plt.plot(zcr3[0], linestyle = 'solid');
plt.suptitle("Zero Crossing Rate graphs for the number 9")
plt.show();
```

We can see that all the three audio files for number 5 have zero crossing rate in the range of (.03, 0.19).

We can see that all the three audio files for number 9 have zero crossing rate in the range of (.01, 0.10).

```python
#energy (rms)
fig, ax = plt.subplots(1, 3,figsize=(15,4))
fig.tight_layout(pad=5.0)

energy1 = librosa.feature.rms(y=y1)
plt.subplot(1, 3, 1)
plt.plot(energy1[0], linestyle = 'solid');

energy2 = librosa.feature.rms(y=y2)
plt.subplot(1, 3, 2)
plt.plot(energy2[0],linestyle = 'solid');

energy3 = librosa.feature.rms(y=y3)
plt.subplot(1, 3, 3)
plt.plot(energy3[0], linestyle = 'solid');
plt.suptitle("Energy graphs for the number 5")
plt.show();
```

```python
fig, ax = plt.subplots(1, 3,figsize=(15,4))
fig.tight_layout(pad=5.0)

energy1 = librosa.feature.rms(y=x1)
plt.subplot(1, 3, 1)
plt.plot(energy1[0], linestyle = 'solid');

energy2 = librosa.feature.rms(y=x2)
plt.subplot(1, 3, 2)
plt.plot(energy2[0],linestyle = 'solid');

energy3 = librosa.feature.rms(y=x3)
plt.subplot(1, 3, 3)
plt.plot(energy3[0], linestyle = 'solid');
plt.suptitle("Energy graphs for the number 9")
plt.show();
```

The major difference we get to see with these graphs is that the graphs of number 5 have sharp peaks while those of number 9 don't. Also, the max energy of number 5 exceeds 0.03 while that of number 9 is limited to that of 0.02.

```python
#spectral roll-off
fig, ax = plt.subplots(1, 3,figsize=(15,4))
fig.tight_layout(pad=5.0)

S, phase = librosa.magphase(librosa.stft(y1))
rolloff = librosa.feature.spectral_rolloff(S=S, sr=sr1, roll_percent=0.85)
plt.subplot(1, 3, 1)
plt.plot(rolloff[0],linestyle = 'solid')

S, phase = librosa.magphase(librosa.stft(y2))
rolloff = librosa.feature.spectral_rolloff(S=S, sr=sr2, roll_percent=0.85)
plt.subplot(1, 3, 2)
plt.plot(rolloff[0],linestyle = 'solid')

S, phase = librosa.magphase(librosa.stft(y3))
rolloff = librosa.feature.spectral_rolloff(S=S, sr=sr3, roll_percent=0.85)
plt.subplot(1, 3, 3)
plt.plot(rolloff[0],linestyle = 'solid')
plt.suptitle("Spectral Roll-off graphs for the number 5")
plt.show()
```

```python
#spectral roll-off
fig, ax = plt.subplots(1, 3,figsize=(15,4))
fig.tight_layout(pad=5.0)

S, phase = librosa.magphase(librosa.stft(x1))
rolloff = librosa.feature.spectral_rolloff(S=S, sr=sr_1, roll_percent=0.85)
plt.subplot(1, 3, 1)
plt.plot(rolloff[0],linestyle = 'solid')

S, phase = librosa.magphase(librosa.stft(x2))
rolloff = librosa.feature.spectral_rolloff(S=S, sr=sr_2, roll_percent=0.85)
plt.subplot(1, 3, 2)
plt.plot(rolloff[0],linestyle = 'solid')

S, phase = librosa.magphase(librosa.stft(x3))
rolloff = librosa.feature.spectral_rolloff(S=S, sr=sr_3, roll_percent=0.85)
plt.subplot(1, 3, 3)
plt.plot(rolloff[0],linestyle = 'solid')
plt.suptitle("Spectral Roll-off graphs for the number 9")
plt.show()
```

```python
#spectral flux
fig, ax = plt.subplots(1, 3,figsize=(15,4))
fig.tight_layout(pad=5.0)

onset_env = librosa.onset.onset_strength(y=y1, sr=sr1)
plt.subplot(1, 3, 1)
plt.plot(onset_env)

onset_env = librosa.onset.onset_strength(y=y2, sr=sr2)
plt.subplot(1, 3, 2)
plt.plot(onset_env)

onset_env = librosa.onset.onset_strength(y=y3, sr=sr3)
plt.subplot(1, 3, 3)
plt.plot(onset_env)
plt.suptitle("Spectral flux graphs for the number 5")
plt.show()
```

**Conclusion:**
---
The following features can be used for differentiating between the number 5 and number 9.

1.   ***Zero Crossing Rate:*** The number 5 has zcr in the range (0.03,0.19) and the number 9 has zcr in the range (0.01,0.10). Also, the values for number 9 increase gradually, reach a maxima and then decrease. But for the number 5, the graph is abrupt and does not have features like the number 9.
2.   ***Energy:*** The graphs of number 5 have sharp peaks while those of number 9 don't. Also, the max energy of number 5 exceeds 0.03 while that of number 9 is limited to that of 0.02.
3.  ***Spectral Roll-off:*** The spectral roll-off values for the number 5 are in the range of (1000,3500) with a maximum value in the range (3000,3500). The spectral roll-off values for the number 9 are in the range of (1000,3000).
4.  ***Pitch:*** The number 5 has pitch values in the range (0,3900), while 9 in the range (0,3400).
